```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook

plt.style.use("ggplot")
```

```python
df = pd.read_csv("screener.csv")
```

```python
used = list(df.iloc[:,3])
br_count = 0
l_count = 0
s_count = 0
n_count = 0
p_count = 0
for response in used:
    if "BeReal" in response:
        br_count += 1
    if "Locket" in response:
        l_count += 1
    if "Spam" in response:
        s_count += 1
    if "Notelt" in response:
        n_count += 1
    if "Paparazzi" in response:
        p_count += 1
```

```python
responses_count = len(used)
sizes = np.array([s_count, l_count, p_count, n_count, br_count])/responses_count*100
labels = ["Spam", "Locket", "Paparazzi", "NoteIt", "BeReal"]
explode = (0,0.05,0,0,0.05)
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=260)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Which social media apps have you heard of?")
plt.savefig("response_usage.png")
plt.show()
```

```python
usage_br_unproc = list(df.iloc[:,5])
usage_br = [x for x in usage_br_unproc if np.isnan(x) == False]

plt.figure()
plt.hist(usage_br)
plt.xlabel("Usage of BeReal by BeReal Users")
plt.ylabel("Number of Respondents")
plt.title("How would you characterize your usage of BeReal?")
plt.savefig("bereal_usage.png")
```

```python
usage_l_unproc = list(df.iloc[:,9])
usage_l = [x for x in usage_l_unproc if np.isnan(x) == False]

plt.figure()
plt.hist(usage_l)
plt.xlabel("Usage of Locket by Locket Users")
plt.ylabel("Number of Respondents")
plt.title("How would you characterize your usage of Locket?")
plt.savefig("locket_usage.png")
```

```python
like_more = list(df.iloc[:,15])
br_count_lm = 0
l_count_lm = 0
n_count_lm = 0
for response in like_more:
    if "BeReal" in response:
        br_count_lm += 1
    if "Locket" in response:
        l_count_lm += 1
    else:
        n_count_lm += 1
```

```python
sizes_lm = np.array([l_count_lm, n_count_lm, br_count_lm])/responses_count*100
labels_lm = ["Locket", "Never Heard of Either", "BeReal"]
explode_lm = (0,0,0.05)
```

```python
fig2, ax2 = plt.subplots()
ax2.pie(sizes_lm, explode=explode_lm, labels=labels_lm, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Which of BeReal or Locket do you like more?")
plt.savefig("like_more.png")
plt.show()
```

```python
more_popular_list = list(df.iloc[:,17])
more_popular = []
for x in more_popular_list:
    if isinstance(x, str):
        more_popular.append(x)
```

```python
br_count_mp = 0
l_count_mp = 0
for response in more_popular:
    if "BeReal" in response:
        br_count_mp += 1
    if "Locket" in response:
        l_count_mp += 1
```

```python
sizes_mp = np.array([l_count_mp, br_count_mp])/(br_count_mp+l_count_mp)*100
labels_mp = ["Locket", "BeReal"]
explode_mp = (0,0.05)
```

```python
fig3, ax3 = plt.subplots()
ax3.pie(sizes_mp, explode=explode_mp, labels=labels_mp, autopct='%1.1f%%',
        shadow=True, startangle=0)
ax3.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Which of BeReal or Locket do you think is more popular?")
plt.savefig("more_popular.png")
plt.show()
```
